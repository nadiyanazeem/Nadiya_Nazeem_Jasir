# Linear Regression Model

```python
x=df.drop(["Price"],axis=1) ## independent variable
y = df['Price']## dependent Variable
x
```

```python
y
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state= 42) 
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
model = lin_regressor.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print ('MSE =',mean_squared_error(y_test,predictions))
print('R squared value = ',r2_score(y_test,predictions))
```

# Lasso Regression`

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
ls_model=lasso.fit(x_train,y_train)
```

```python
ls_predict=ls_model.predict(x_test)
```

```python
print("Mean Squared Error=",mean_squared_error(y_test,ls_predict))
print("R Squared Error=",r2_score(y_test,ls_predict))
```
